**Las tuplas, a diferencia de las listas, no se pueden modificar. Podemos convertir una tupla en lista, editarla, y volver a convertirla en tupla (NO modificable).**

```python
# Creamos una tupla
tupla_nombres = ("Lola", "Paula")

# Cambiamos tupla a lista
lista_nombres = list(tupla_nombres)

print(lista_nombres)
```

```python
# Añadimos un nombre más a la lista eligiendo en qué posición, por ello usamos insert
lista_nombres.insert(2,"Andrea")
print(lista_nombres)
```
